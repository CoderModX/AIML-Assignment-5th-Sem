```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
x = df.iloc[:,[2,3]].values
```

```python
x
```

```python
x.shape
```

```python
X = pd.DataFrame(x)
db = DBSCAN(eps=8,min_samples=9, metric='euclidean')
model = db.fit(X)
labels  = model.labels_
no_c = len(set(labels)) - (1 if -1 in labels else 0)
print(no_c)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=labels)
```

```python
db2 = DBSCAN(eps=12.75,min_samples=10, metric='euclidean')
model2 = db2.fit(x)
labels2  = model2.labels_
no_c2 = len(set(labels2)) - (1 if -1 in labels2 else 0)
print(no_c2)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=labels2)
```

```python
db3 = DBSCAN(eps=0.25,min_samples=3, metric='euclidean')
model3 = db3.fit(x)
labels3  = model3.labels_
no_c3 = len(set(labels3)) - (1 if -1 in labels3 else 0)
print(no_c3)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=labels3)
```
