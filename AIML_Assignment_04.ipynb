# Matplotlib

```python
male = df['Sex'].value_counts()['male']
female = df['Sex'].value_counts()['female']
fare = df['Fare'].mean()

bar_name = ['Male','Female', 'Mean of Fare']

count = [male,female,fare]
color = ['coral','gold','deepskyblue']
```

```python
fig = plt.figure(figsize = (10,7))
plt.bar(bar_name,count,color=color)
plt.show()
```
