j) The zero (0) values presents in "Glucose", "BloodPressure", "SkinThickness" and "Insulin"
attributes replace with ‘NaN’ and then fill all NaN entries with corresponding attribute’s 
mean value.

```python
cols = ["GLUCOSE", "BLOODPRESSURE","SKINTHICKNESS","INSULIN"]
df[cols] = df[cols].replace({'0':np.nan,0:np.nan})
df.head()
```

```python
def replace_nan(df,column):
    m = np.mean(df[column])
    m = round(m,2)
    df[column].fillna(value = m , inplace = True)
```

```python
for i in cols:
    replace_nan(df,i)
```
